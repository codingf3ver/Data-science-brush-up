```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

#### Descriptive Statistics (Customer Spending Analysis)
1. Spending Amount : The amount a customer spends per visit.
2. Mean (Average Spending): The central value of customer spending.
3. Median (Middle Spending): The middle spending value, useful when data is skewed.
4. Mode (Most Frequent Spending): The spending amount that occurs the most.
5. Variance (Spread of Spending): Measures how far spending values deviate from the mean.
6. Standard Deviation (Consistency in Spending): High value means customers spend very differently; low means consistent spending.

```python


# Sample customer spending data (in rupees)
spending = [50, 75, 100, 120, 150, 200, 200, 220, 300, 500]

# Mean, Median, Mode
mean_value = np.mean(spending)
median_value = np.median(spending)
mode_value = stats.mode(spending).mode

# Variance & Standard Deviation
variance_value = np.var(spending, ddof=1)
std_dev_value = np.std(spending, ddof=1)

print(f"Mean: {mean_value}, Median: {median_value}, Mode: {mode_value}")
print(f"Variance: {variance_value}, Standard Deviation: {std_dev_value}")

# Visualization
plt.figure(figsize=(8,5))
sns.histplot(spending, bins=5, kde=True, color='blue')
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label="Mean")
plt.axvline(median_value, color='green', linestyle='dashed', linewidth=2, label="Median")
plt.axvline(mode_value, color='purple', linestyle='dashed', linewidth=2, label="Mode")
plt.legend()
plt.title("Customer Spending Distribution")
plt.show()
```

#### Inferential Statistics (A/B Testing)

1. Dataset Attributes and Their Meaning:
2. Conversion Rate (%): Percentage of website visitors who make a purchase.
3. Old Website Performance: Conversion rates from the current version.
4. New Website Performance: Conversion rates after a redesign.
5. T-Statistic (Effect Strength): Measures how different the two website performances are.
6. p-Value (Significance Level): If below 0.05, the new website has a statistically significant improvement.


```python
from scipy.stats import ttest_ind

# Conversion rates 
old_conversion = [5, 7, 8, 6, 7, 7, 6, 5, 7, 6]
new_conversion = [8, 9, 7, 10, 9, 10, 8, 9, 10, 8]

# T-test
t_stat, p_value = ttest_ind(new_conversion, old_conversion)
print(f"T-Statistic: {t_stat}, p-Value: {p_value}")

# Visualization
plt.figure(figsize=(6, 4))
sns.boxplot(data=[old_conversion, new_conversion])
plt.xticks([0, 1], ["Old Website", "New Website"])
plt.title("Conversion Rate Distribution")
plt.ylabel("Conversion Rate")
plt.show()
```

#### Probability Theory (Fraud Detection in Banking)

1. Dataset Attributes and Their Meaning:
2. Fraud Probability (P_Fraud): The likelihood a transaction is fraudulent.
3. Normal Transaction Probability (P_Normal): The chance a transaction is legitimate.
4. Alert Accuracy (P_Alert_Fraud): The system's ability to flag fraud correctly.
5. False Alarm Rate (P_Alert_Normal): How often normal transactions are wrongly flagged.
6. Bayes Theorem Output: The probability that an alert is a true fraud case.


```python
# Given probabilities
P_Fraud = 0.02  # Probability of fraud
P_Normal = 1 - P_Fraud
P_Alert_Fraud = 0.90  # Alert when fraud occurs
P_Alert_Normal = 0.05  # False alert

# Bayes' Theorem
P_Alert = (P_Alert_Fraud * P_Fraud) + (P_Alert_Normal * P_Normal)
P_Fraud_Alert = (P_Alert_Fraud * P_Fraud) / P_Alert

print(f"Probability of Fraud given an Alert: {P_Fraud_Alert:.2f}")

# Visualization
categories = ['Fraud Detected', 'False Alarm']
values = [P_Fraud_Alert, 1 - P_Fraud_Alert]

plt.figure(figsize=(6, 4))
sns.barplot(x=categories, y=values, palette="coolwarm")
plt.title("Fraud Detection Probability")
plt.show()
```

#### Regression Analysis (House Price Prediction)

1. Square Footage (X): The size of the house in square feet.
2. House Price (Y): The sale price of the house in dollars.
3. Regression Line: A model that predicts price based on size.
4. Prediction: Forecasts price for unseen house sizes.

```python
from sklearn.linear_model import LinearRegression

# Sample house data (Square footage, Price)
X = np.array([1200, 1500, 1700, 2000, 2200, 2500]).reshape(-1, 1)
y = np.array([250000, 270000, 290000, 310000, 330000, 350000])

# Train model
model = LinearRegression()
model.fit(X, y)

# Predictions
predicted = model.predict(X)

# Visualization
plt.figure(figsize=(8,5))
plt.scatter(X, y, color='blue', label="Actual Prices")
plt.plot(X, predicted, color='red', linewidth=2, label="Regression Line")
plt.xlabel("Square Footage")
plt.ylabel("House Price ($)")
plt.title("House Price Prediction")
plt.legend()
plt.show()
```

#### Clustering (Customer Segmentation)
1. Customer Spending ($): Total amount a customer spends.
2. Visits Per Month: Number of times the customer shops monthly.
3. Clusters: Groups of similar customers (e.g., Low, Medium, High spenders).

```python
from sklearn.cluster import KMeans

# Customer spending data (Spending amount, Visits per month)
customer_data = np.array([[50, 2], [200, 5], [300, 7], [20, 1], [500, 10], [100, 3]])

# Train K-Means
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(customer_data)
labels = kmeans.labels_

# Visualization
plt.figure(figsize=(6, 6))
plt.scatter(customer_data[:, 0], customer_data[:, 1], c=labels, cmap='viridis', s=100)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='red', marker='X', s=200, label="Centroids")
plt.xlabel("Spending ($)")
plt.ylabel("Visits per Month")
plt.title("Customer Segmentation (K-Means Clustering)")
plt.legend()
plt.show()
```

#### Time Series Analysis (Stock Market Prediction)

```python
import statsmodels.api as sm

# Sample stock prices
stock_prices = [120, 122, 124, 125, 130, 128, 135, 140, 138, 145]

# Fit ARIMA model
model = sm.tsa.ARIMA(stock_prices, order=(2, 1, 2))
result = model.fit()

# Predict next 3 days
forecast = result.forecast(steps=3)

# Visualization
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), stock_prices, marker='o', label="Actual Prices")
plt.plot(range(11, 14), forecast, marker='s', linestyle='dashed', color='red', label="Forecasted Prices")
plt.xlabel("Time (Days)")
plt.ylabel("Stock Price ($)")
plt.title("Stock Price Prediction")
plt.legend()
plt.show()
```
