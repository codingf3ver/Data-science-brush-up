##### 1. Classes and objects

```python
# Instance mthods

"""âœ… 1. Instance Method

The most common type of method.

Works with object/instance data.

Must have self as the first parameter.

Can access instance variables and class variables."""

class Person:
    def __init__(self, name):
        self.name = name  
    def show(self):     
        return f"My name is {self.name}"
```

```python
p = Person("Alice")
print(p.show())
```

```python
# class methods
"""âœ… 2. Class Method

Works with class-level data, not with individual object data.

Has cls as the first parameter.

Can access only class variables, not instance variables.

Use the decorator @classmethod."""
class Person:
    count = 0  

    def __init__(self):
        Person.count += 1
        self.count2=4

    @classmethod
    def show_count(cls):  
        return f"Total persons: {cls.count}"
```

```python
Person.show_count()
```

```python
"""âœ… 3. Static Method

Does not take self or cls.

Behaves like a normal function inside a class.

Cannot access instance variables or class variables.

Use @staticmethod decorator.

Used for utility/helper functions that relate to the class."""

class MathOps:
    @staticmethod
    def add(a, b):
        return a + b
```

```python
MathOps.add(5, 3)
```

##### 2. Inheritance in python

```python
# parent class
# single inheritance
class Car:
    def __init__(self,windows,door,engine_type):
        self.windows = windows
        self.door = door
        self.engine_type = engine_type
    
    def drive(self):
        print(f'person will drive the car : {self.engine_type}')
```

```python
car1 = Car(4,2,'petrol')
car1.drive()
```

```python
class Tesla(Car):
    def __init__(self, windows, door, engine_type,is_self_driving):
        super().__init__(windows, door, engine_type) # inheriting from car class
        self.is_self_driving = is_self_driving
    
    def self_driving(self):
        print(f'tesla support self driving : {self.is_self_driving}')
        print(f'this car type is :  {self.engine_type}')
```

```python
tesla1 = Tesla(4,4,'electric',True)
```

```python
tesla1.self_driving()
```

```python
tesla1.drive()
```

```python
# Multiple inheritance
# when a class inherita property from more than one class

# Base class 1
class Aninal:
    def __init__(self,name):
        self.name = name

    def speak(self):
        return "subclass must inherit"

# Base class 2
class Pet:
    def __init__(self,owner):
        self.owner = owner

    def color_city(self):
        return f'{self.owner} is from kolkata'

# derived class
class Dog(Aninal,Pet):
    def __init__(self, name,owner):
        Aninal.__init__(self,name)
        Pet.__init__(self,owner)

    def speak(self):
        return f'{self.name} says woof from derived : {super().speak()} : {super().color_city()} '

```

```python
dog  = Dog("Buddy",'tausif')
```

```python
dog.speak()
```

##### 3. Polymorphism

```python
# mehtod overiding

class Animal:
    def speak(self):
        return "sound of the animal"
    
class Dog(Animal):
    def speak(self):
        return "the dog makes a sound woof"
    
class Cat(Animal):
    def speak(self):
        return "Meow"

def animal_speak(animal):
    print(animal.speak())
```

```python
dog = Dog()
cat  = Cat()
# print(dog.speak())
# print(cat.speak())
animal_speak(dog)
animal_speak(cat)
```

```python
### Polymorphism with Functions and Methods
## base class
class Shape:
    def area(self):
        return "The area of the figure"
    
## Derived class 1
class Rectangle(Shape):

    def __init__(self,width, height):
        self.width=width
        self.height=height
    
    def area(self):
        return self.width*self.height

## Derived class 2
class Circle(Shape):
    def __init__(self,radius):
        self.radius  = radius

    def area(self):
        return 3.14*self.radius*self.radius

def print_area(shape):
    print(shape.area())
```

```python
rectangle = Rectangle(4,5)
circle = Circle(4)

print_area(rectangle)
print_area(circle)
```

####  Polymorphism with Abstract Base Classes
##### Abstract Base Classes (ABCs) are used to define common methods for a group of related objects. They can 
##### enforce that derived classes implement particular methods, promoting consistency across different
##### implementations.

```python
from abc import ABC, abstractclassmethod

class Vehicle(ABC):
    @abstractclassmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        return "car engine has started"
    
class Motorcycle(Vehicle):
    def start_engine(self):
        return "motor cycle engine has started"
    
def print_start_engine(vehicle):
    print(vehicle.start_engine())
```

```python
car = Car()
motor = Motorcycle()
print_start_engine(car)
print_start_engine(motor)
```

##### 4. Encapsulation
Encapsulation is the concept of wrapping data (variables) and methods (functions) together as a single unit. 
It
restricts direct access to some of the object's components, which is a means of preventing accidental
interference and misuse of the data.

```python
# encapsulation with getter and setter
# public, private and protected

class Person:
    def __init__(self,name,age):
        self.name = name #public variables
        self.age = age #public variables

def get_name(person):
    return person.name


person = Person("tausif",30)
print(get_name(person))
# print(person.name)
```

##### 5. Abstraction

```python
"""@abstractmethod is a decorator from the abc module used to define a method without implementation.

Forces subclasses to implement the method

Ensures a common structure

Used to achieve abstraction and polymorphism"""
```

```python
from abc import ABC, abstractmethod

# Abstract class
class Payment(ABC):

    @abstractmethod
    def pay(self, amount):
        pass

# Subclass implementing abstract method
class CreditCardPayment(Payment):

    def pay(self, amount):
        return f"Paid â‚¹{amount} using Credit Card"

class UpiPayment(Payment):

    def pay(self, amount):
        return f"Paid â‚¹{amount} using UPI"

# Using the classes
payment1 = CreditCardPayment()
print(payment1.pay(500))

payment2 = UpiPayment()
print(payment2.pay(300))

# ðŸ”Ž Explanation

# Payment is an abstract class â†’ you cannot create an object of it.

# pay() is an abstract method â†’ no body, just a definition.

# Every subclass must implement pay() â†’ this enforces structure.

# This hides internal logic and exposes only the necessary interface.
```
