#### 1. difference b/w for and yeild

```python
"""1. for loop
Executes all iterations immediately.
Stores all values in memory (if iterating over a list or building a list).
Used for regular iteration."""
```

```python
def get_numbers():
    numbers = []
    for i in range(5):
        numbers.append(i)
    return numbers

```

```python
print(get_numbers())
```

```python
"""
2. yield (generator)

Used inside a function to create a generator.
Produces values one at a time, only when needed (lazy evaluation).
Saves memory — does not store the full result.
Very useful for big data, streams, pipelines.
"""
```

```python
def get_numbers():
    for i in range(5):
        yield i
```

```python
print(get_numbers())
```

```python
for num in enumerate(get_numbers()):
    print(num)
```

```python
# basics magic methods
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def __str__(self):
        return f"{self.name} is {self.age} years old"

    def __repr__(self):
        return f"person (name= {self.name}) (age = {self.age})"

person = Person("Tausif",30)
```

#### 3. method overloading

```python
'''
Common Operator Overloading Magic Methods
__add__(self, other): Adds two objects using the + operator.
__sub__(self, other): Subtracts two objects using the operator.
__mul__(self, other): Multiplies two objects using the * operator.
__truediv__(self, other): Divides two objects using the / operator.
__eq__(self, other): Checks if two objects are equal using the operator.
__lt__(self, other): Checks if one object is less than another using the < operator.
'''
```

```python
class Vector:
    
    def __init__(self, x, y ):
        self.x = x
        self.y = y
    
    def __add__(self,other):
        return Vector(self.x + other.x ,self.y + other.y)

    def __sub__(self,other):
        return Vector(self.x - other.x ,self.y - other.y)

    def __mul__(self,other):
        return Vector(self.x * other ,self.y * other)

    def __eq__(self,other):
        return self.x == other.x and self.y == other.y

    def __repr__(self):
        return f"Vector({self.x},{self.y})"

```

```python
v1 = Vector(4,6)
v2 = Vector(5,7)
```

```python
print(v1+v2)
print(v1-v2)
print(v1*3)
```

```python
class ComplexNumber:
    def __init__(self, real=0, imag=0):
        self.real = real
        self.imag = imag

    def __str__(self):
        return f"{self.real} + {self.imag}i"

    # Operator Overloading: +
    def __add__(self, other):
        if isinstance(other, ComplexNumber):
            return ComplexNumber(self.real + other.real,
                                 self.imag + other.imag)

        # Overloading: adding integer or float
        elif isinstance(other, (int, float)):
            return ComplexNumber(self.real + other, self.imag)
        else:
            return NotImplemented


    def __sub__(self, other):
        if isinstance(other, ComplexNumber):
            return ComplexNumber(self.real - other.real,
                                 self.imag - other.imag)
        elif isinstance(other, (int, float)):
            return ComplexNumber(self.real - other, self.imag)
        else:
            return NotImplemented

    def __mul__(self, other):
        if isinstance(other, ComplexNumber):
            # (a + bi)(c + di) = (ac − bd) + (ad + bc)i
            real = self.real * other.real - self.imag * other.imag
            imag = self.real * other.imag + self.imag * other.real
            return ComplexNumber(real, imag)
        elif isinstance(other, (int, float)):
            return ComplexNumber(self.real * other, self.imag * other)
        else:
            return NotImplemented

```

```python
c1 = ComplexNumber(4, 5)
c2 = ComplexNumber(2, 3)
```

```python
print("Addition       : ", c1 + c2)
print("Subtraction    : ", c1 - c2)
print("Multiplication : ", c1 * c2)
```

```python
print("Add integer    : ", c1 + 10)
print("Multiply int   : ", c1 * 2)
```

#### 4. Custom exception handling

```python
class Error(Exception):
    pass

class dobException(Error):
    pass
```

```python
year = int(input('enter your age year: '))
age = 2025 - year

try:
    if age <=30 and age>=20:
        print('The age is valid')
    else:
        raise dobException
except dobException:
    print('Your age is not valid must be between 30 and 20')
```

#### Decorators

```python
def main_function(func):
    
    def inner_function():
        print('test inner 1')
        func()
        print(f'Main variable access from inner ')
    return inner_function()
```

```python
def outer_function():
    print('outer function')
```

```python
main_function(outer_function)
```

```python
@main_function
def outer_function():
    print('outer function using decorator')
```

```python
def my_decorator(func):

    def wrapper():
        print('start from inside function')
        func()
        print('end from inside function')
    return wrapper
```

```python
@my_decorator
def outer_func_for_deco():
    print('hello from outside')
outer_func_for_deco()
```

```python
# decorator wit argumwnts
```

```python
def repeat(n):
    def my_decorator(func):
        def wrapper(*args,**kwargs):
            for _ in range(n):
                func(*args,**kwargs)
        return wrapper
    return my_decorator
```

```python
@repeat(5)
def outer_function_test():
    print('hello from outside')
```

```python
outer_function_test()
```
